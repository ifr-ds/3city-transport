## ZTM Gdańsk

### load data

```python
ztm_gdansk_url = "https://ckan.multimediagdansk.pl/dataset/c24aa637-3619-4dc2-a171-a23eec8f2172/resource/4c4025f0-01bf-41f7-a39f-d156d201b82b/download/stops.json"

with urllib.request.urlopen(ztm_gdansk_url) as url:
    data = json.loads(url.read().decode())
```
