## Heatmap without time

```python
from folium import plugins
from folium.plugins import HeatMap

map_3city = folium.Map(location=[54.441564, 18.562754], zoom_start = 13) 

df['stopLat'] = df['stopLat'].astype(float)
df['stopLon'] = df['stopLon'].astype(float)

heat_df = df[['stopLat', 'stopLon']].rename(columns={"stopLat": "Latitude", "stopLon": "Longitude"})

# Create weight column
heat_df['Weight'] = df['time_sec']
heat_df['Weight'] = heat_df['Weight'].astype(float)
heat_df = heat_df.dropna(axis=0, subset=['Latitude','Longitude', 'Weight'])

# List comprehension to make out list of lists
heat_data = [[row['Latitude'],row['Longitude']] for index, row in heat_df.iterrows()]

# Plot it on the map
HeatMap(heat_data).add_to(map_3city)

# Display the map
map_3city
```
